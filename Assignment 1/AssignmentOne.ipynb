### Assignment 1: Temperature Check

```python
temp = int(input("Enter current temperature: "))

if temp < 10:
    print("It's too cold")
elif temp > 10 and temp < 25:
    print("It's cold outside")
else:
    print("It's hot outside")
```

### Assignment 2: ATM Withdrawal System

```python
balance = int(input("Enter Your Balance: "))
withdraw = int(input("Enter Your Withdraw Balance: "))
if withdraw <= balance:
    print("Transaction Successful!")
    print("Now Current Balance:", balance - withdraw)

else:

    print("Insufficient Balance!")
```

### Assignment 3: Student Grade Calculator

```python
mark = int(input("Enter your marks: "))

if mark <= 100 and mark >= 90:
    print("You Got A+")
elif mark <= 89 and mark >= 80:
    print("You Got A")
elif mark <= 79 and mark >= 70:
    print("You Got B")
elif mark <= 69 and mark >= 60:
    print("You Got C")
else:
    print("You Failed")


```

### Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter Your Age to know about Bus Ticket Discount: "))

if age < 5:
    print("Bus Ticket Free")
elif age >= 5 and age <= 18:
    print("You Got 50% Discount")
elif age >= 60:
    print("You Got 30% Discount")
else:
    print("You have to pay the full price of the ticket.")

```

### Assignment 5: Restaurant Bill with Discount

```python
totalBill = float(input("Input the total bill amount"))

if totalBill > 1000:
    totalBill = totalBill - (totalBill * 10)/100
    print("You Got 10% Discount. The final bill  amount: ",totalBill )
else:
    print("The final bill  amount:",totalBill )
```

### Assignment 7: Traffic Light System

### Assignment 8: Leap Year Checker

```python
year = int(input("Enter a Year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not Leap Year")

```

### Assignment 10: Electricity Bill Calculator 


```python
units = int(input("Enter number of units consumed: "))

if units >= 0 and units <= 100:
    bill = units * 5
elif units >= 101 and units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print("Total payable amount:", bill)
```
