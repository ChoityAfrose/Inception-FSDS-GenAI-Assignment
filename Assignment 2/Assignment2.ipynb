### 1. Email Validator (String Filtering + Loops)

```python
email = input("Enter your email: ") # Input: hello@inceptionbd.com
valid = True 

# No spaces allowed, Must contain @, Only one @ allowed
if " " in email:
    valid = False
elif "@" not in email:
    valid = False
elif email.count("@") != 1:
    valid = False
else:
    pass

if valid == True:
    at_index = email.index("@")
    first_part = email[0:at_index]
    second_part = email[at_index+1:len(email)]

    special_characters = "!#$%^&*()-_=+[]{}|;:'\",.<>/?`~"

    # first part must not be empty
    if len(first_part) == 0:
        valid = False
    # cannot start or end with special characters
    elif first_part[0] in special_characters or first_part[-1] in special_characters:
        valid = False

    elif len(second_part) == 0:
        valid = False
    
    elif second_part[0] in special_characters:
        valid = False

    elif not (email.endswith(".com") or email.endswith(".org")):
        valid = False

    elif ".." in email:
        valid = False
    else:
        pass

# Final result
if valid == True:
    print("Valid Email")
else:
    print("Invalid Email")




```

### 2. Count Word Frequency in Customer Reviews

```python
text_str = input("Enter a review or comment: ") # Input: This product is good and this service is excellent
text_str2 = text_str.lower()
words = text_str2.split(" ")

print("{", end="")

for i in range(len(words)):

    if words[i] in words[:i]: #skip if the word already appeared before
        continue

    count = words.count(words[i]) #count how many times the current word appears

    if words[i] == words[len(words)-1]:
          print(f"'{words[i]}': {count}", end = "")
    else:
        print(f"'{words[i]}': {count}", end = ", ")
        
print("}")
```

### 3. Password Strength Checker

```python
has_upper = False
has_lower = False
has_digit = False
has_special = False

special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>/?`~"
is_strong = False

while is_strong == False:
    password = input("Enter a password: ") # Input: Hello@123

# Loop through each character
    for i in password:
        if i.isupper():
            has_upper = True
        elif i.islower():
            has_lower = True
        elif i.isdigit():
            has_digit = True
        elif i in special_characters:
            has_special = True
        else:
            pass
   
    is_long = len(password) >= 8

# Final Result
    if has_upper and has_lower and has_digit and has_special and is_long:
        print("Strong Password")
        is_strong = True
    else:
        print("Weak Password")
        if not has_upper:
            print("- Add at least one uppercase letter (A-Z)")
        if not has_lower:
            print("- Add at least one lowercase letter (a-z)")
        if not has_digit:
            print("- Add at least one number (0-9)")
        if not has_special:
            print("- Add at least one special character (!,@,#,$,%, etc.)")
        if not is_long:
            print("- Password must be at least 8 characters long")
```

### 4. Username Generator

```python
import random

full_name = input("Enter full name: ") # Input: Mamun Malitha
name_parts = []
word = ""

# Split full name manually using loop
for ch in full_name:
    if ch == " ":
        if word != "":
            name_parts.append(word)
            word = ""
    else:
        word += ch
if word != "":
    name_parts.append(word)

first_name = name_parts[0]
last_name = name_parts[-1]

count = 0
first_part = ""
last_part = ""

for i in first_name:
    if count < 3:
        count += 1
        first_part +=i

length = len(last_name)
        
for i in range(length-3,length):
    idx = 0
    for ch in last_name:
        if idx == i:
            last_part += ch
        idx += 1

# Generate random number (2-digit)
rand_num = random.randint(10, 99)

user_name = first_part.lower() + last_part.lower() + str(rand_num)

print("Generated Username:", user_name)
```

### 5. SMS Word Counter

```python
sms_word = input("Enter a demo message: ") # Input: This is a demo message.

words = 0
in_word = False
characters = 0
status = ""

for ch in sms_word:

    if ch != " " and in_word == False:
        words += 1
        in_word = True
    elif ch == " ":
        in_word = False
    else: 
        pass

    characters += 1

if characters <= 160:
    status = "Within Limit"
else:
    status = "Exceed Limit"

print (f"Words = {words}, Characters = {characters}, Status = {status}")

```

### 6. Remove Duplicates from Contact List

```python
contact_str = "Rahim, Karim, Rahim, Rahim, Sultana"
contact_list = []
word = ""

for ch in contact_str:
    if ch == ",":
        if word.strip() != "":
            contact_list.append(word.strip())

        word = ""
    else:
        word +=ch

if word.strip() != "":
    contact_list.append(word.strip())

unique_list = []

for  contact in contact_list:
    already_exists = False

    for un in unique_list:
        if un == contact:
            already_exists = True
            break

    if already_exists == False:
        unique_list.append(contact)

output = ""

for un in range(len(unique_list)):
    output += unique_list[un]
    
    if un != len(unique_list) - 1:
        output += ", "
        
print('Output: "' + output + '"')
```

### 7. Detect Suspicious Words in Text (Moderation)

```python
banned_words = ['spam', 'scam', 'fake']
found = False

text = input("Enter content: ") # input : This is a scam offer
lower_text = text.lower()
word = ""
text_list = []

for ch in lower_text:
    if ch != " ":
        word += ch
    elif ch == " " and word != "":
        text_list.append(word.strip())
        word = ""
#print(text_list)

# Check if any banned word exists
for text in text_list:
    for bw in banned_words:
        if text == bw:
            found = True
            break
    if found :
        break

if found:
    print(f"Warning: Suspicious content detected.")
else:
    print("Safe: No banned words found")
```

### 8. Invoice Code Formatter

### 10. Word Reversal for Data Encryption

```python
my_string = input("Enter a sentence: ") #AI is fun
reversed_str = ""
reversed_word = ""
word = ""
reversed_list = []

for ch in my_string:
    if ch != " ":
        word += ch
    elif ch == " " and word != "":
        reversed_word = word[::-1]
        reversed_list.append(reversed_word)
        word = ""
        reversed_word = ""
    else:
        pass

if word != "":
    reversed_word = word[::-1]
    reversed_list.append(reversed_word)

result = ""
for i in range(len(reversed_list)):
    result += reversed_list[i]
    if i != len(reversed_list) - 1:
        result += " "
print("Reversed Words Sentence:", result)
```
