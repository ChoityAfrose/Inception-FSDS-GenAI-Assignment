## 1. Inventory Price Calculator

```python
def calculate_inventory_summary (prices):

    # if list is empty
    if not prices:
        return 0, 0 

    total_cost = 0

    for  i in prices:
        total_cost += i  

    average_cost = total_cost/len(prices)
    return total_cost, average_cost


# Price example 120 250 399 150 
p = input("Enter prices separated by space: ").split() 
price = [float(x) for x in p]
result = calculate_inventory_summary (price)
print(tuple(result))
```

## 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitors =sorted(set(visitor_list))  # Unique visitor ID
    num_unique_visitors = len(unique_visitors) # Number of unique visitor
    return num_unique_visitors, unique_visitors

# Visitors ID example 101 205 101 310 205 550
v = input("Enter visitors ID separated by space: ").split() 
visitor = [int(x) for x in v]
count, visitors = count_unique_visitors(visitor)

print(f"{count} (The unique IDs are ", end = "")

for x in range(len(visitors)):
    if x == len(visitors) - 1:
        print(f"{visitors[x]})") # last one 
    else:
        print(f"{visitors[x]}, ", end = "")
```

## 3. Product Category Counter

```python
def count_categories(products):

    category_count = {}
    
    for product in products:
        if product in category_count:
            category_count[product] += 1 #increase product
        else:
            category_count[product] = 1  #add new product

    return category_count

# Example = "Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"
L = input("Enter products separated by space: ").split()
product_list = [i for i in L]

# Call function 
result = count_categories(product_list)
print(result)  
```

## 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_temp = []
    normal_temp = []
    for i in temp_list:
        if i > 38:
            high_temp.append(i)
        else:
            normal_temp.append(i)

    return high_temp, normal_temp

t = input("Enter temperature separated by space: ").split() # [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9] 
temp = [float(x) for x in t]
result = analyze_temperatures(temp) #function call
print(tuple(result))
```

## 5. Menu Price Lookup

```python
def get_price(my_dict,my_item):
    
    for key, value in my_dict.items():
        if key.lower() == my_item.lower():
            return value
    
    return "item not found"
    
my_dict = {}

n = int(input("Enter how many food do you add"))

# Example menu dict for input {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
for i in range(n):
    key = input("Enter product name: ")
    value = float(input("Enter price of this food: "))
    my_dict[key] = value

print("Welcome our Restaurant. This is our menu: ")

for item, price in my_dict.items():
    print(f"{item} : {price}")

my_item = input("Enter your food name for Order: ")
result  = get_price(my_dict,my_item)
print(f"{my_item} food item price is {result}")
```

## 6. Student Grading System

```python
def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    elif score >= 50:
        return "B"
    elif score >= 40:
        return "C"
    elif score >= 33:
        return "D"
    else:
        return "F"
    
def grade_students(student_dict):
    
    std_grade_dict = {}

    for key, value in student_dict.items():
        grade = calculate_grade(value)
        std_grade_dict[key] = grade 
        
    return std_grade_dict


# Example Students score dictionary {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
n = int(input("Enter how many students do you want to add: "))

for i in range(n):
    key = input("Enter Student name: ")
    value = float(input("Enter Score of the Student: "))
    my_dict[key] = value

result = grade_students(my_dict)
print(result)
```

## 7. Discount Calculator

```python
def apply_discounts(products):

    discount_list = []
    
    for name, price, discount in products:

        updated_price = ((100 - discount) * price)/100
        discount_list.append(updated_price)

    return discount_list

# Example products[("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
products = []
n = int(input("How many products? "))

for i in range(n):
    name = input("Enter product name: ")
    price = float(input("Enter price: "))
    discount = float(input("Enter discount: "))
    
    products.append((name, price, discount))

print(products)
result = apply_discounts(products)
print(result)
```

## 8. Hashtag Generator

```python
import string
def generate_hashtags(sentence):

    hash_tag = []
    words = sentence.split()

    for i in words:
        word = i.strip(string.punctuation) #Strip punctuation
        hashtag_word = "#"+word
        hash_tag.append(hashtag_word)

    return hash_tag

sen = input("Enter a Sentence: ")
result = generate_hashtags(sen.lower())
print(result)
```

## 10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
    unique_items = set(cart)
    total_amount = 0

    for item in unique_items:
        if item in prices:
            quantity = cart.count(item)
            total_amount += prices[item] * quantity
        else:
            print(f"Warning: '{item}' not found in price list!")

    return total_amount  

number = int(input("How many number of product do you add? "))
price = {}

# Example dictionary {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}
for i in range(number):
    key = input("Enter product Name: ")
    value = float(input("Enter product Price: "))
    price[key] = value

print(price)

input_cart = input("Enter your products: ") # Milk Bread Milk Cheese Bread
cart = input_cart.split()
print(cart)

result = calculate_total(cart, price)
print("Total amount:",result)
```
